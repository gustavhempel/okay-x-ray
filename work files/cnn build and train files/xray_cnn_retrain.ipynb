```
import pandas as pd
import numpy as np
from PIL import Image
from sklearn.metrics import confusion_matrix
from keras.preprocessing.image import ImageDataGenerator, array_to_img
from google.colab import drive
from tensorflow import keras

#set name for model
file_name = "ver3_model_softmax_3L64u_3Lnode_5epochs"

#connect to google drive
drive.mount('/content/gdrive')
root_path = 'gdrive/My Drive/xray_files/'  #change dir to your project folder

```

```

model = keras.models.load_model(f'{root_path}{file_name}.h5') 
```

```
hyper_dimension = 500
hyper_epochs = 5
hyper_batch_size = 6
hyper_feature_maps = 32
hyper_channels = 1
hyper_mode = 'grayscale'
```

```
# Creating training and validation image flows

train_gen = ImageDataGenerator(rescale = 1./255,
                               shear_range = 0.2,
                               zoom_range = 0.2,
                               vertical_flip = True)

val_gen = ImageDataGenerator(rescale = 1./255)

train_set = train_gen.flow_from_directory(f'{root_path}train',
                                          target_size = (hyper_dimension,
                                                         hyper_dimension),
                                          batch_size = hyper_batch_size,
                                          class_mode = 'categorical',
                                          color_mode = hyper_mode)

val_set = val_gen.flow_from_directory(f'{root_path}val',
                                      target_size = (hyper_dimension,
                                                     hyper_dimension),
                                      batch_size = hyper_batch_size,
                                      class_mode = 'categorical',
                                      color_mode = hyper_mode)
```

```
# Creating test image flow

test_gen = ImageDataGenerator(rescale = 1./255)

test_set = test_gen.flow_from_directory(f'{root_path}/test',
                                        target_size = (hyper_dimension,
                                                       hyper_dimension),
                                        batch_size = 1,
                                        class_mode = 'categorical',
                                        color_mode = hyper_mode,
                                        shuffle=False)

# Making predictions

```

```
predictions = model.predict(test_set)
```

```
predictions
```

```
y_classes = predictions.argmax(axis=-1)
```

```
# Confusion matrix

cm = pd.DataFrame(data=confusion_matrix(test_set.classes, y_classes, labels=[0, 1]),
                  index=["Actual Normal", "Actual Pneumonia"],
                  columns=["Predicted Normal", "Predicted Pneumonia"])
cm
```
